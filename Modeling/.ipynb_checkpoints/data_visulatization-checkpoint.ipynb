```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import PIL
import torch
import torchvision.transforms as T

from skimage.transform import rotate, AffineTransform, warp
from skimage.util import random_noise
from skimage.filters import gaussian

import random

import dask.array as da
```

Image augmentation is only applied to training set to generalize the model prediction.

```python
task= 4 # 8
if task == 4:
        #label: 1: steel, 0: other
        org_X_train_data_dir = 'data/task4_X_train.npy'
        org_Y_train_data_dir = 'data/task4_y_train.npy'
        num_classes = 2
        claases = ['other', 'steel']
elif task == 8:
        # label: 0: combined damage, 1: Flexural damage, 2: Undamaged state, 3: Shear damage 
        org_X_train_data_dir = 'data/task8_X_train.npy'
        org_Y_train_data_dir = 'data/task8_y_train.npy'
        num_classes = 4
        claases = ['comb', 'flex', 'ud', 'shear']
```

```python
task = 4
x_train_dataset = np.load(org_X_train_data_dir)
y_train_dataset = np.load(org_Y_train_data_dir) 
```

Image Normalization

```python
# Finding mean and var for normalization
# task = 4
(N, W, H, rgb) = x_train_dataset.shape
r_sum = 0
g_sum = 0
b_sum = 0
total = N*W*H
r_sum_sq = 0
g_sum_sq = 0
b_sum_sq = 0
for n in range(N):
  d = x_train_dataset[n]
  r_sum += np.sum(d[:, :, 0])
  g_sum += np.sum(d[:, :, 1])
  b_sum += np.sum(d[:, :, 2])
  d = x_train_dataset[n]**2
  r_sum_sq += np.sum(d[:, :, 0])
  g_sum_sq += np.sum(d[:, :, 1])
  b_sum_sq += np.sum(d[:, :, 2])

r_avg = r_sum/total
g_avg = g_sum/total
b_avg = b_sum/total

r_var = (r_sum_sq/total-r_avg**2)
g_var = (g_sum_sq/total-g_avg**2)
b_var = (b_sum_sq/total-b_avg**2)
print("Mean:", r_avg, g_avg, b_avg, "\n STD:", r_var, g_var, b_var)
```

## Image Visualization

```python
# in raw format
plt.imshow(x_train_dataset[0], interpolation="nearest")  
if all(y_train_dataset[0] == np.array([1., 0.])):
  header = "Label = Other"
else:
  header = "Label = Steel"
plt.title(header)
plt.show()
```

```python
# after changing to original colors
ind = 0
d = x_train_dataset[ind]
data_0 = np.zeros(d.shape)
for rgb in range(3):
    data_0[:,:,rgb] = (d[:,:,rgb]-d[:,:,rgb].min())*255/(d[:,:,rgb].max()-d[:,:,rgb].min())
# plotting
PIL.Image.fromarray(data_0.astype(np.uint8))
```

### Gray-Scale Image

```python
# oversample label = 1 as it has less 
img = Image.fromarray(data_0.astype(np.uint8))
gray_img = T.Grayscale()(img) # this 224*224 (not 3*224*224) We may not be able to use it withe same data
gray_img_np = np.array(gray_img)
PIL.Image.fromarray(gray_img_np.astype(np.uint8))
# plt.imshow(gray_img, interpolation="nearest")  
```

### Normalize

```python
img = Image.fromarray(data_0.astype(np.uint8))
norm_img = T.Normalize(mean=(23.036, 15.335, 12.189), std=(60.694, 58.007, 58.368))(T.ToTensor()(img))  # 3*224*224
norm_img = norm_img.permute(1,2,0)
norm_img_np = np.array(norm_img)
PIL.Image.fromarray(norm_img_np.astype(np.uint8))
```

### Rotation

```python
img = Image.fromarray(data_0.astype(np.uint8))
rotated_img = T.RandomRotation(degrees=250)(img)
rotated_img_np = np.array(rotated_img)
PIL.Image.fromarray(rotated_img_np.astype(np.uint8))
```

```python
# alternatively in np
rotated_img_np = rotate(data_0, angle=45, mode = 'wrap')
PIL.Image.fromarray(rotated_img_np.astype(np.uint8))
```

### Gaussian blur

```python
img = Image.fromarray(data_0.astype(np.uint8))
blurred_img = T.GaussianBlur(kernel_size=(51, 91), sigma=2)(img)
blurred_img_np = np.array(blurred_img)
PIL.Image.fromarray(blurred_img_np.astype(np.uint8))
```

```python
# alternatively (all done in np)
blurred_img = gaussian(data_0, sigma=2, multichannel=True)
PIL.Image.fromarray(blurred_img.astype(np.uint8))
```

### flipping

```python
# LR flip
flip_lr_img = np.fliplr(data_0) #numpy array type
PIL.Image.fromarray(flip_lr_img.astype(np.uint8))
```

```python
#UD flip
flip_ud_img = np.flipud(data_0) #numpy array type
PIL.Image.fromarray(flip_ud_img.astype(np.uint8))
```

## Adding Noise

```python
sigma=0.1
noisy_img = random_noise(data_0, var=sigma**2)
# PIL.Image.fromarray(noisy_img.astype(np.uint8))

plt.imshow(noisy_img)
plt.title('Random Noise')
```
